```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import MaxAbsScaler
```

```python
def cost(x_train,y_train,m):
    return (1/x_train.shape[0])*((y_train-(x_train*m).sum(axis=1))**2).sum(axis=0)
```

```python
def stepGradient(m, alpha,x_train, y_train):
    m_slopes = np.zeros(len(m))
    N = x_train.shape[0]
    M = x_train.shape[1]
    
    for i in range(N):
        for j in range(M):
            m_slopes[j] += (-2/N)*(y_train[i] - (m*x_train[i]).sum())*x_train[i][j]
    m = m - alpha*m_slopes
    return m 
```

```python
def gd(x_train,y_train,alpha,itr):
    m_final=np.zeros(x_train.shape[1])
    cst=np.zeros(itr)
    #print(x_train)
    for i in range(itr):
        m_final=stepGradient(m_final,alpha,x_train,y_train)
        cst[i]=cost(x_train, y_train, m_final)
        print("Cost", i+1, cst[i])
    return m_final,cst
```

```python
def predict2(x_test,m):
    print(m)
    return ((m[:-1]*x_test).sum(axis=1)+m[-1])
```

```python
data=np.loadtxt('0000000000002419_training_ccpp_x_y_train.csv',delimiter=',')
x_test=np.loadtxt('0000000000002419_test_ccpp_x_test.csv',delimiter=',')
x_train=data[:,:-1]
print(x_train.shape,x_test.shape)
```

```python
temp=np.ones((x_train.shape[0],x_train.shape[1]*4))
temp[:,:x_train.shape[1]]=x_train
temp[:,x_train.shape[1]:x_train.shape[1]*2]=x_train**2
temp[:,x_train.shape[1]*2:x_train.shape[1]*3]=x_train**3
temp[:,x_train.shape[1]*3:x_train.shape[1]*4]=x_train**4
x_train=temp


temp=np.ones((x_test.shape[0],x_test.shape[1]*4))
temp[:,:x_test.shape[1]]=x_test
temp[:,x_test.shape[1]:x_test.shape[1]*2]=x_test**2
temp[:,x_test.shape[1]*2:x_test.shape[1]*3]=x_test**3
temp[:,x_test.shape[1]*3:x_test.shape[1]*4]=x_test**4
x_test=temp
```

```python
print(x_train.shape,x_test.shape)
```

```python
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
Z = np.ones((len(x_train[:, 0]), len(x_train[0]) + 1))
Z[:,:-1] = x_train
x_train = Z

y_train=data[:,-1]

m,cst=gd(x_train,y_train,0.1,5000)
```

```python
y_pred=predict2(x_test,m)
y_pred
```
